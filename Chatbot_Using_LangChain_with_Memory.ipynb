```python
config = {"configurable": {"session_id": "firstchat"}}
```

```python
store={}
```

```python
def get_session_history(session_id:str) -> BaseChatMessageHistory:
  if session_id not in store:
    store[session_id]=InMemoryChatMessageHistory()
  return store[session_id]
```

```python
model_with_memory=RunnableWithMessageHistory(model,get_session_history)
```

```python
model_with_memory.invoke([HumanMessage(content="Hi! I'm satyam")],config=config)
```

```python
model_with_memory.invoke([HumanMessage(content="Hi! what is my name")],config=config)
```

```python
from langchain_core.prompts import ChatPromptTemplate, MessagesPlaceholder
```

```python
prompt = ChatPromptTemplate.from_messages(
    [
        ("system", "you are a helpful assistant. Answer all questions to the best of you ability."),
        MessagesPlaceholder(variable_name="messages"),

    ]
)
```

```python
chain = prompt | model
```

```python
chain.invoke({"messages": [HumanMessage(content="hi! I am satyam")]})
```

```python
chain.invoke({"messages": [HumanMessage(content="what is my name")]}).content
```

```python
model_with_memory=RunnableWithMessageHistory(chain,get_session_history)
```

```python
config = {"configurable": {"session_id": "secondChat"}}
```

```python
model_with_memory.invoke([HumanMessage(content="hi! I am satyam")],config=config)
```

```python
response=model_with_memory.invoke([HumanMessage(content="what is my name")],config=config)
```

```python
from langchain_core.messages import SystemMessage, trim_messages
```

```python
trimmer= trim_messages(
    max_tokens=40,
    strategy="last",
    token_counter=model,
    include_system=True,
    allow_partial=False,
    start_on="human",
)

```

```python
messages=[
    HumanMessage(content="hi! I am satyam"),
    AIMessage(content="Hello satyam! How can I assist you today"),
    HumanMessage(content="I love cricket"),
    AIMessage(content="nice"),
    HumanMessage(content="what is 2+2"),
    AIMessage(content="4"),
    HumanMessage(content="thanks"),
    AIMessage(content="you're welcome")
]
```

```python
messages
```

```python
messages+[HumanMessage(content="what is my name")]
```

```python
model.get_num_tokens_from_messages(messages)
```

```python
trimmer.invoke(messages)
```

```python
trimmed_messages=trimmer.invoke(messages)
```

```python
model.get_num_tokens_from_messages(trimmed_messages)
```

```python
from operator import itemgetter
from langchain_core.runnables import RunnablePassthrough
chain=(
    RunnablePassthrough.assign(messages=itemgetter("messages") | trimmer)
    | prompt
    | model
)

response = chain.invoke({
    "messages": messages+[HumanMessage(content="what is my name")],
    "language":"English"
})
```

```python
response.content
```

```python
model_with_memory=RunnableWithMessageHistory(
    chain,get_session_history,
    input_messages_key="messages",


)
config={"configurable": {"session_id":"thirdChat"}}
```

```python
response=model_with_memory.invoke(
    {
    "messages":messages+[HumanMessage(content="what is my name")],
    "language":"English"
    },
    config=config

)
response.content
```
